Cálculo das estatísticas básicas do atributo idade usando as funções do Pandas.

```python
quantidade = df['Idade'].count()
print('Quantidade de elementos: ', quantidade)

soma = df['Idade'].sum()
print('Soma dos elementos: ', soma)

media = df['Idade'].mean()
print('Media dos elementos: ', media)

mediana = df['Idade'].median()
print('Mediana dos elementos: ', mediana)

quartil1 = df['Idade'].quantile(q=0.25)
print('Quartil 1 dos elementos: ', quartil1)

quartil2 = df['Idade'].quantile(q=0.50)
print('Quartil 2 (mediana) dos elementos: ', quartil2)

quartil3 = df['Idade'].quantile(q=0.75)
print('Quartil 3 dos elementos: ', quartil3)

moda = df['Idade'].mode()
print('Moda dos elementos: ')
for x in moda:
  print(x)

minimo = df['Idade'].min()
print('Menor valor: ', minimo)

maximo = df['Idade'].max()
print('Maior valor: ', maximo)

var = df['Idade'].var()
print('Variância dos elementos: ', var)

dp = df['Idade'].std()
print('Desvio padrão dos elementos: ', dp)

cv = dp / media * 100
print('Coeficiente de variação dos elementos: ', cv)

assimetria = df['Idade'].skew()
print('Coeficiente de assimetria: ', assimetria)

curtose = df['Idade'].kurtosis()
print('Coeficiente de Curtose: ', curtose)
```

**Segunda parte**

Criação do histograma de frequência para a variável Idade, usando a fórmula de Sturges, usando a função hist. Neste exemplo, a série **idades** foi criada a partir de uma das colunas do dataframe.



```python
idades = df['Idade']
#Fórmula de Sturges
classes = int(1 + np.log2(idades.count()))

plt.hist(idades,bins=classes, rwidth=0.99)
plt.title('Distribuição de Frequência por Idades')
plt.xlabel('Idade')
plt.ylabel('Frequência')
plt.show()
```

É possível criar os gráficos diretamente, usando funções do dataframe que chamam as funções da bibliteca matplotlib.

```python
df['Idade'].hist(bins=classes,rwidth=0.95)
plt.title('Distribuição de Frequência por Idades')
plt.xlabel('Idade')
plt.ylabel('Frequência')
plt.show()
```

Ao passar o valor 'auto' para o parâmetro bins, o histograma será criado com a quantidade de classes maior, entre a fórmula de Sturges e a de Freedman-Diaconis.

```python
df['Idade'].hist(bins='auto',rwidth=0.95)
plt.title('Distribuição de Frequência por Idades')
plt.xlabel('Idade')
plt.ylabel('Frequência')
plt.show()
```

```python
sns.histplot(data=df, x='Idade', bins=classes, shrink=0.95)
plt.title('Distribuição de frequência por idade')
plt.ylabel('Frequência')
plt.show()
```

```python
sns.histplot(data=df, x="Idade", binwidth=10, shrink=0.95)
plt.title('Distribuição de frequência por idade')
plt.ylabel('Frequência')
plt.show()
```

**Terceira Parte**

Cálculo da covariância e da correlação entre as variáveis quantitativas do Dataframe.

Ao calcular o coeficiente de correlação deve-se indicar o tipo de cálculo a ser feito, por meio do parâmetro method, que pode receber os seguintes valores:


*   pearson (padrão);
*   kendall;
*   spearman.

```python
df.cov(numeric_only=True)
```

```python
df.corr(numeric_only=True)
```

```python
df.corr(method='kendall',numeric_only=True)
```

```python
df.corr(method='spearman',numeric_only=True)
```

```python
corr = df.corr(numeric_only=True)
sns.heatmap(corr,annot=True)
plt.show()
```

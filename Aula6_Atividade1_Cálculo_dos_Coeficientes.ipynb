```python
import numpy as np
import matplotlib.pyplot as plt

cores = ['b','g','r','c','m','y','k']
legenda=[]
legenda.append('Pontos reais')
x = np.array([2,4,6])
y = np.array([5,9,13])
lr = 0.01
plt.title('Cálculo dos coeficientes de regressão linear simples')
plt.scatter(x,y)

w0 = 0
w1 = 3

for i in range(7):
  y_pred = w1*x + w0
  legenda.append(f'Reta {w1:.3f}x +  {w0:.3f}')
  plt.plot(x,y_pred,c=cores[i])
  erro = y - y_pred
  soma = erro.sum()
  w0 = w0 + lr*soma
  soma2 = (erro*x).sum()
  w1 = w1 + lr*soma2

plt.legend(legenda)
plt.show()
```
